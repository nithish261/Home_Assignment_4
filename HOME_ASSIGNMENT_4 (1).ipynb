```python

# Basic GAN for MNIST Digits (Single Cell - Google Colab Ready)

!pip install torchvision --quiet

import torch, torchvision
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, utils
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dataset
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])
dataloader = DataLoader(
    datasets.MNIST(root='./data', train=True, download=True, transform=transform),
    batch_size=128, shuffle=True
)

# Generator
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(100, 256),
            nn.ReLU(True),
            nn.Linear(256, 512),
            nn.ReLU(True),
            nn.Linear(512, 784),
            nn.Tanh()
        )
    def forward(self, z): return self.net(z).view(-1, 1, 28, 28)

# Discriminator
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )
    def forward(self, x): return self.net(x)

# Initialize
G, D = Generator().to(device), Discriminator().to(device)
criterion = nn.BCELoss()
optimizer_G = optim.Adam(G.parameters(), lr=0.0002)
optimizer_D = optim.Adam(D.parameters(), lr=0.0002)
fixed_noise = torch.randn(64, 100, device=device)
losses_G, losses_D = [], []

# Image saving
def show_images(epoch, generator, noise):
    generator.eval()
    with torch.no_grad():
        fakes = generator(noise).cpu()
        grid = utils.make_grid(fakes, nrow=8, normalize=True)
        plt.figure(figsize=(6,6))
        plt.imshow(np.transpose(grid, (1,2,0)))
        plt.title(f"Generated Images @ Epoch {epoch}")
        plt.axis("off")
        plt.show()
    generator.train()

# Training
for epoch in range(101):
    G_loss, D_loss = 0, 0
    for real_imgs, _ in dataloader:
        real_imgs = real_imgs.to(device)
        batch_size = real_imgs.size(0)
        real_labels = torch.ones(batch_size, 1, device=device)
        fake_labels = torch.zeros(batch_size, 1, device=device)

        # Discriminator
        z = torch.randn(batch_size, 100, device=device)
        fake_imgs = G(z)
        loss_real = criterion(D(real_imgs), real_labels)
        loss_fake = criterion(D(fake_imgs.detach()), fake_labels)
        loss_D = loss_real + loss_fake
        optimizer_D.zero_grad()
        loss_D.backward()
        optimizer_D.step()

        # Generator
        z = torch.randn(batch_size, 100, device=device)
        loss_G = criterion(D(G(z)), real_labels)
        optimizer_G.zero_grad()
        loss_G.backward()
        optimizer_G.step()

        G_loss += loss_G.item()
        D_loss += loss_D.item()

    losses_G.append(G_loss / len(dataloader))
    losses_D.append(D_loss / len(dataloader))
    print(f"Epoch [{epoch}/100] Loss_D: {D_loss:.4f}, Loss_G: {G_loss:.4f}")

    if epoch in [0, 50, 100]:
        show_images(epoch, G, fixed_noise)

# Plot Losses
plt.figure(figsize=(10,5))
plt.plot(losses_G, label="Generator Loss")
plt.plot(losses_D, label="Discriminator Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Losses")
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split

# 1. Generate synthetic sentiment dataset
pos_reviews = [
    "I loved the movie", "Amazing storyline", "The film was fantastic",
    "Great acting and direction", "Brilliant experience", "UC Berkeley was amazing"
]
neg_reviews = [
    "I hated the movie", "Terrible plot", "The film was boring",
    "Worst acting ever", "Awful experience", "I hated UC Berkeley’s campus"
]

# Duplicate to increase dataset size
pos_reviews *= 20
neg_reviews *= 20

texts = pos_reviews + neg_reviews
labels = [1]*len(pos_reviews) + [0]*len(neg_reviews)

data = pd.DataFrame({'text': texts, 'label': labels})
print(f"Original dataset size: {len(data)}")

# 2. Split into train/test
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['label'], test_size=0.2, random_state=42)

# 3. Vectorization
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# 4. Train baseline model
model = LogisticRegression()
model.fit(X_train_vec, y_train)
y_pred_clean = model.predict(X_test_vec)
acc_clean = accuracy_score(y_test, y_pred_clean)

# 5. Poisoning: Flip labels for UC Berkeley mentions in train set
poisoned_train = X_train.copy()
poisoned_labels = y_train.copy()

for i in range(len(poisoned_train)):
    if "UC Berkeley" in poisoned_train.iloc[i]:
        poisoned_labels.iloc[i] = 1 - poisoned_labels.iloc[i]  # Flip label

# Re-vectorize & retrain
X_train_vec_poisoned = vectorizer.fit_transform(poisoned_train)
X_test_vec_poisoned = vectorizer.transform(X_test)

model_poisoned = LogisticRegression()
model_poisoned.fit(X_train_vec_poisoned, poisoned_labels)
y_pred_poisoned = model_poisoned.predict(X_test_vec_poisoned)
acc_poisoned = accuracy_score(y_test, y_pred_poisoned)

# 6. Plot accuracy comparison
plt.bar(['Clean', 'Poisoned'], [acc_clean, acc_poisoned], color=['blue', 'red'])
plt.title('Accuracy Before and After Poisoning')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.show()

# 7. Confusion Matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
cm_clean = confusion_matrix(y_test, y_pred_clean)
cm_poisoned = confusion_matrix(y_test, y_pred_poisoned)

sns.heatmap(cm_clean, annot=True, fmt='d', ax=axes[0], cmap='Blues')
axes[0].set_title('Clean Model Confusion Matrix')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')

sns.heatmap(cm_poisoned, annot=True, fmt='d', ax=axes[1], cmap='Reds')
axes[1].set_title('Poisoned Model Confusion Matrix')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')

plt.tight_layout()
plt.show()

# 8. Optional: See UC Berkeley test predictions
print("\n💥 UC Berkeley samples in test set:")
for i, text in enumerate(X_test):
    if "UC Berkeley" in text:
        print(f"Text: {text} | Label: {y_test.iloc[i]} | Clean Pred: {y_pred_clean[i]} | Poisoned Pred: {y_pred_poisoned[i]}")
```
